```python
col = ["id", "id_info", "نماد", "نام نماد", "?", "اولین", "پایانی", "معامله", "تعداد معاملات", "حجم معاملات", "ارزش معاملات", "کمترین بازه روز", "بیشترین بازه روز",
       "بازده دیروز", "EPS", "حجم مبنا", "?", "?", "کد گروه صنعت", "قیمت مجاز بیشترین", "قیمت مجاز کمترین", "تعداد سهام", "?", "NAV ابطال", "موقعیت های باز"]

df = df_split.copy()

for i in range(len(col)):
    df.rename(columns={i:f"{col[i]}"}, inplace=True)

df
```

```python
df_ektiar = df[df["نام نماد"].apply(lambda x:"اختيار" in x)].reset_index(drop=True)
# df_ektiar
df_ektiar.to_excel(r"../tsetmc/Data/df_ektiar.xlsx")
df_ektiar
```

```python
df_split_values= df_ektiar["نام نماد"].str.split('-')
df_ektiar["نام نماد"] = df_split_values.str[0]
df_ektiar["مقدار"] = df_split_values.str[1]
df_ektiar["تاریخ"] = df_split_values.str[2]
df_ektiar.insert(3, "نام نماد", df_ektiar.pop("نام نماد"))
df_ektiar.insert(4, "مقدار", df_ektiar.pop("مقدار"))
df_ektiar.insert(5, "تاریخ", df_ektiar.pop("تاریخ"))

#df_ektiar = df_ektiar.drop(columns=["نام نماد"])
# df_ektiar.to_excel(r"..\tsetmc\df_ektiar.xlsx")
df_ektiar
```

```python
def convert_time(time):
    if len(time) == 8 and '/' not in time:
        year = int(time[:4])
        month = int(time[4:6])
        day = int(time[6:])
        return f"{year}/{month:02d}/{day:02d}"
    else:
        return time

# def convert_year(time):
#     time_split = time.split('/')
#     if len(time_split) == 6:
#         year = int(time_split[:2])
#         month = int(time_split[2:4])
#         day = int(time_split[4:])
#         return f"14{year}/{month:02d}/{day:02d}"
#     else:
#         return time
   
df_ektiar['تاریخ'] = df_ektiar['تاریخ'].apply(convert_time)
# df_ektiar['تاریخ'] = df_ektiar['تاریخ'].apply(convert_year)

df_ektiar
```
